```
from sklearn.feature_extraction.text import TfidfVectorizer
```

```
import pandas as pd
```

```
f=open('a.txt','r')
```

```
x=f.read()
ls=x.split('.')
ls
```

```
ls.pop()
```

```
import nltk
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_wd=set(stopwords.words('english'))
#stop_wd
```

```
vectorizer = TfidfVectorizer()
```

```
vectors = vectorizer.fit_transform(ls)
```

```
feature_names = vectorizer.get_feature_names()
```

```
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
```

```
df
```

```
import numpy as np
X = np.array(df)
from sklearn.decomposition import NMF
model = NMF(n_components=None, init='random', random_state=0)
W = model.fit_transform(X)
H = model.components_
```

```
print(W)
W.shape
```

```
w=pd.DataFrame.from_records(W,columns=feature_names)
w
```

```
l=w.mean()
wd=l.to_dict()
wd
```

```
len(wd)
```

```
from heapq import nlargest 
mwt = nlargest(20, wd, key = wd.get)
mwt 
```

```
imp_wd=[i for i in mwt if i not in stop_wd ]
imp_wd
```

```
import nltk
tokenizer=nltk.tokenize.TreebankWordTokenizer()
```

```
sen_wt={}
for n,i in enumerate(ls):
  d=0
  words=tokenizer.tokenize(i)
  for j in words:
    if j in imp_wd:
      d=d+wd[j]
  sen_wt[n]=d

sen_wt
```

```
s = nlargest(8, sen_wt, key =sen_wt.get)
s
```

```
s.sort()
s
```

```
st=''
for i in s:
  st+=''.join(ls[i])+'.'

st
```
